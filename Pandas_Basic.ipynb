# Basic Pandas - Index

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
# dataframe is essential a two dimentiaonl array - data dictionry 
df = pd.DataFrame(people)
df
```

```python
df.set_index('email')
# pass the column that I would like to set as index 

df
# when look at dataframe again, it is still as the old one. as panda doesnt change the data 
# unless you specify in place

```

```python
df.set_index('email', inplace = True)

df
# set email column as index, and set this one in place change, now the index is the label of this row 
```

```python
df.loc['JohnDoe@email.com']
# pass the index value to identify the row
```

```python
df.reset_index(inplace = True)
df
# reset the index and so its back to origin 
```

# Basic Pandas - Filter data

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
# dataframe is essential a two dimentiaonl array - data dictionry 
df = pd.DataFrame(people)
```

```python
df['last'] == 'Doe'  
# mask filter - it return whether the row is true or not
```

```python
# way 1
filt = (df['last']=='Doe')
# create a filter statement 
df[filt]
df.loc[filt]
```

```python
# way 2
df[df['last']=='Doe']
```

```python
df.loc[filt,'email']
# loc[rows we want, colum we want]
# we can pass this filt true, false into loc, it will return the matching rows
```

```python
filt = (df['last'] == 'Doe') & (df['first'] =='John')

df.loc[filt, 'email']

# two condition as filter AND
```

```python
filt = (df['last'] == 'Doe') | (df['first'] =='John')

df.loc[filt]

# two condition as filter OR 
```
